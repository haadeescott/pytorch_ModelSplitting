```python
import pandas as pd
import matplotlib.pyplot as plt
```

# **Alexnet Results**



> Plots


1.   CPU Utilization (Gramine SGX)
2.   CPU Utilization (Native)
3.   Memory Footprint (Gramine SGX)
4.   Memory Footprint (Native)
5.   CPU Utilization for all Splits (1/2/4/6/8/10/12) (Gramine SGX)

7.   Power Consumption (Gramine SGX)

```python
df = pd.read_csv('Alexnet_metrics.csv', usecols=alexnet_cols)
```

```python
plt.rcParams["figure.figsize"] = [9.50, 3.50]
plt.rcParams["figure.autolayout"] = True
```

```python
plt.plot(df.Time, df.Cpu_Utilization)
plt.title('Alexnet - 1 model')
plt.xlabel('Time in second(s)')
plt.ylabel('CPU Utilization (%)')
plt.ylim([0, 100])
plt.savefig('plot_Images/alexnet_cpuUtil.png')
plt.show()
```

```python
plt.plot(df.Time, df.Cpu_Util_NOSGX)
plt.title('Alexnet - 1 model')
plt.xlabel('Time in second(s)')
plt.ylabel('CPU Utilization (%)')
plt.ylim([0, 10])
plt.savefig('plot_Images/alexnet_cpuUtil_NOSGX.png')
plt.show()
```

```python
plt.plot(df.Time, df.Memory_Footprint)
plt.title('Alexnet - 1 model')
plt.xlabel('Time in second(s)')
plt.ylabel('Memory Footprint (KB)')
plt.ylim([200000, 2000000])
plt.savefig('plot_Images/alexnet_MemF.png')
plt.show()
```

```python
plt.plot(df.Time, df.Mem_F_NOSGX)
plt.title('Alexnet - 1 model')
plt.xlabel('Time in second(s)')
plt.ylabel('Memory Footprint (KB)')
plt.ylim([200000, 2000000])
plt.savefig('plot_Images/alexnet_MemF_NOSGX.png')
plt.show()
```

```python
plt.plot(df.Time, df.Cpu_Utilization, df.Time, df.Cpu_Util_2,
         df.Time, df.Cpu_Util_2,
         df.Time, df.Cpu_Util_4,
         df.Time, df.Cpu_Util_6,
         df.Time, df.Cpu_Util_8,
         df.Time, df.Cpu_Util_10,
         df.Time, df.Cpu_Util_12)
plt.title('Alexnet - All models')
plt.xlabel('Time in second(s)')
plt.ylabel('CPU Utilization (%)')
plt.ylim([0, 100])
plt.legend(['1 Model', '2 Submodels', '4 Submodels', '6 Submodels', '8 Submodels', '10 Submodels', '12 Submodels'], loc='upper left')
plt.savefig('plot_Images/alexnet_cpuUtil_All.png')
plt.show()
```

```python
plt.plot(df.Time, df.Memory_Footprint, df.Time, df.Mem_F_2,
         df.Time, df.Mem_F_4,
         df.Time, df.Mem_F_6,
         df.Time, df.Mem_F_8,
         df.Time, df.Mem_F_10,
         df.Time, df.Mem_F_12)
plt.title('Alexnet - All models')
plt.xlabel('Time in second(s)')
plt.ylabel('Memory Footprint (KB)')
plt.ylim([200000, 2000000])
plt.legend(['1 Model', '2 Submodels', '4 Submodels', '6 Submodels', '8 Submodels', '10 Submodels', '12 Submodels'], loc='upper right')
plt.savefig('plot_Images/alexnet_MemF_All.png')
plt.show()
```

```python
plt.plot(df.Power_Time, df.Power_Consumption)
plt.title('Alexnet - 1 model')
plt.xlabel('Time in second(s)')
plt.ylabel('Power Consumption (W)')
plt.ylim([0, 10])
plt.savefig('plot_Images/alexnet_PowerConsumption.png')
plt.show()
```

```python
plt.plot(df.Power_Time, df.Power_Consumption,
         df.Power_Time, df.Power_Consumption_2,
         df.Power_Time, df.Power_Consumption_4,
         df.Power_Time, df.Power_Consumption_6,
         df.Power_Time, df.Power_Consumption_8,
         df.Power_Time, df.Power_Consumption_10,
         df.Power_Time, df.Power_Consumption_12,)
plt.title('Alexnet - ALL models')
plt.xlabel('Time in second(s)')
plt.ylabel('Power Consumption (W)')
plt.ylim([0, 10])
plt.legend(['1 Model', '2 Submodels', '4 Submodels', '6 Submodels', '8 Submodels', '10 Submodels', '12 Submodels'], loc='upper right')
plt.savefig('plot_Images/alexnet_PowerConsumption_ALL.png')
plt.show()
```
